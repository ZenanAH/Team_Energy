```python
# Imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# <center>Data Visualization of Energy Data</center>

```python
# Load the dataset

small_data= pd.read_csv('../raw_data/small_data.csv')
small_data
```

```python
# Converting DateTime column to DateTime format

small_data['DateTime'] = pd.to_datetime(small_data['DateTime'])
small_data.dtypes
```

```python
# Look at the first day of data for one house (Class E)
# DataFrame - Group E - 1 day

df_E1D = small_data.head(48)
df_E1D
```

```python
plt.title('One House - Group E - One Day')
plt.plot('DateTime', 'KWH/hh', data = df_E1D);
```

```python
# Looking at longer term trends

df_E = small_data[small_data['LCLid'] == 'MAC000018']
df_E
```

```python
plt.title('One House - Group E - Two Years')
plt.plot('DateTime', 'KWH/hh', data = df_E);
```

```python
# Average Daily consumption for One House - Group E

df_E['KWH/hh'].mean() * 48
```
