```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Creating a Summary of all available data

In advance of extracting and preparing data, we want to first identify if the data is suitable for our purposes.  This had previously been done manually, but we are now automating and looping through the full dataset


```python
# Household information import
# This table provides data for classification we will use

df_hhinfo = pd.read_csv('../raw_data/household_information.csv')
df_hhinfo = df_hhinfo[['LCLid', 'stdorToU','Acorn','Acorn_grouped','file','Groups','Correct_ACORN_group']]
df_hhinfo
```

```python
# Block importing
# There are 111 blocks of data to be iterated through

# Create empty lists for iteration
housename = []
start_date = []
end_date = []
count = []
zeros = []
tariff = []
group = []
classification = []
acorn = []
block = []

for j in range(0, 112):
    # define filepath
    filepath = f'../raw_data/halfhourly_dataset/block_{j}.csv'
    
    # load individual blocks
    df_block = pd.read_csv(filepath)
    
    # Convert DateTime column to DateTime format for keymatching and ease of viewing
    df_block['DateTime'] = pd.to_datetime(df_block['tstp'])
    df_block.drop(columns = 'tstp', inplace = True)
    
    # Remove text strings with 'Null' value in them to allow us to convert energy usage to float
    df_block.drop(df_block[df_block['energy(kWh/hh)'] == 'Null'].index, inplace = True)
    df_block['energy(kWh/hh)'] = df_block['energy(kWh/hh)'].astype(float)
    
    # Store the LCLid's in a string called houselist for each block
    houselist = df_block['LCLid'].unique()
    
    # Status for iteration
    print(f'Now appending block {j} ...')
    
    # Append data from each block to lists to form a new summary dataframe
    for i in range (0,len(houselist)):
        housename.append(houselist[i])
        start_date.append(df_block[df_block['LCLid'] == houselist[i]]['DateTime'].iloc[0])
        end_date.append(df_block[df_block['LCLid'] == houselist[i]]['DateTime'].iloc[-1])
        count.append(df_block[df_block['LCLid'] == houselist[i]].count()[0])
        zeros.append((df_block[df_block['LCLid'] == houselist[i]]['energy(kWh/hh)'] == 0).sum())
        if houselist[i] in df_hhinfo['LCLid'].tolist():
            tariff.append((df_hhinfo[df_hhinfo['LCLid'] == houselist[i]]['stdorToU'].iloc[0]))
            acorn.append((df_hhinfo[df_hhinfo['LCLid'] == houselist[i]]['Acorn'].iloc[0]))
            block.append((df_hhinfo[df_hhinfo['LCLid'] == houselist[i]]['file'].iloc[0]))
            group.append((df_hhinfo[df_hhinfo['LCLid'] == houselist[i]]['Groups'].iloc[0]))
            classification.append((df_hhinfo[df_hhinfo['LCLid'] == houselist[i]]['Correct_ACORN_group'].iloc[0]))
        else:
            tariff.append('No Data')
            acorn.append('No Data')
            block.append('No Data')
            group.append('No Data')
            classification.append('No Data')
```

```python
# Creating a DataFrame with the lists created through iteration

summary = pd.DataFrame({'LCLid': housename, 'start_date':start_date, 'end_date':end_date, 'data_points':count, 'zeros':zeros, 'tariff': tariff, 'Block':block, 'Acorn':acorn, 'Group':group, 'Classification':classification})
```

```python
# Creating two new columns from existing DateTime data

summary['number_of_days'] = (summary['end_date'] - summary['start_date']).dt.days
summary['missing_hh'] = (((summary['end_date'] - summary['start_date']).dt.total_seconds() / 1800) + 1) - summary['data_points']
summary['missing_hh'] = summary['missing_hh'].astype(int)
```

```python
# Reordering columns and viewing final table

summary = summary[['LCLid', 'start_date', 'end_date', 'number_of_days', 'data_points', 'missing_hh', 'zeros', 'tariff', 'Block', 'Acorn', 'Group', 'Classification']]
summary
```

```python
# Exporting dataset as CSV

filepath = '../raw_data/summary_v1.csv'
summary.to_csv(filepath)
```
